```python
def greet_user(name):
  print("hello",(name))
greet_user("sravs")
greet_user("sri")
```

```python
def greet_user(name="sravs"):
  print(f"welcome {name}")

greet_user()
greet_user("ravi")
greet_user("ram")
```

```python
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("jai", "sree", "ram")

```

```python
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="sravs", age=18, city="kmm")
```

```python
def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 3, name="Ravi", city="kmm")
```

```python
def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 4:", square(4))
```
